```python
import pickle
import matplotlib.pyplot as plt
import scipy.stats as stats
import numpy as np

####################################
file_name_1 = 'HalfCheetah_E(4)P_(64,10,100,32,1,0.0001)'
step_num_1 = 5000
label_1 = 'EP ::: BS = 64'

file_name_2 = 'HalfCheetah_E(4)P(8)_(128,10,100,32,1,0.0001)'
step_num_2 = 5000
label_2 = 'EP ::: BS = 128, REMEM = 8'

file_name_3 = 'HalfCheetah_E(4)P(10)_(256,10,100,32,1,0.0001)'
step_num_3 = 5000
label_3 = 'EP ::: BS = 256, REMEM = 10'

file_name_4 = 'HalfCheetah_E(4)P(10)_(64,10,100,32,1,0.0001)'
step_num_4 = 5000
label_4 = 'EP ::: BS = 64, REMEM = 10'
####################################

with open('/home/jolp/Desktop/Data/'+file_name_1+'_seed1', 'rb') as fin1:
    step_avg_list_1 = pickle.load(fin1)
    
with open('/home/jolp/Desktop/Data/'+file_name_2+'_seed1', 'rb') as fin3:
    step_avg_list_2 = pickle.load(fin3)
    
with open('/home/jolp/Desktop/Data/'+file_name_3+'_seed1', 'rb') as fin5:
    step_avg_list_3 = pickle.load(fin5)

with open('/home/jolp/Desktop/Data/'+file_name_4+'_seed1', 'rb') as fin7:
    step_avg_list_4 = pickle.load(fin7)
    
############################################################################

x_1_values = list(range(1, step_num_1 + 1))
x_2_values = list(range(1, step_num_2 + 1))
x_3_values = list(range(1, step_num_3 + 1))
x_4_values = list(range(1, step_num_4 + 1))
y_1_values = step_avg_list_1[:step_num_1]
y_2_values = step_avg_list_2[:step_num_2]
y_3_values = step_avg_list_3[:step_num_3]
y_4_values = step_avg_list_4[:step_num_4]

plt.plot(x_1_values, y_1_values, c='green', label=label_1)
plt.plot(x_2_values, y_2_values, c='red', label=label_2)
plt.plot(x_3_values, y_3_values, c='blue', label=label_3)
plt.plot(x_4_values, y_4_values, c='yellow', label=label_4)

plt.title('HalfCheetah, Epsilon')
plt.grid(True)
plt.legend(loc=2)
plt.show()

print'GRAPH1 : ', label_1
print('    MAX : {}, EPISODE : {}'.format(round(np.max(step_avg_list_1),3),np.argmax(step_avg_list_1)+1))
print('')
print'GRAPH2 : ', label_2
print('    MAX : {}, EPISODE : {}'.format(round(np.max(step_avg_list_2),3),np.argmax(step_avg_list_2)+1))
print('')
print'GRAPH3 : ', label_3
print('    MAX : {}, EPISODE : {}'.format(round(np.max(step_avg_list_3),3),np.argmax(step_avg_list_3)+1))
print('')
print'GRAPH4 : ', label_4
print('    MAX : {}, EPISODE : {}'.format(round(np.max(step_avg_list_4),3),np.argmax(step_avg_list_4)+1))

```
